# **COVID19 Data Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
df = pd.read_csv("Covid_raw_data.csv")
```

```python
df.head()
```

### **Data Cleaning**

```python
# Null Value checking
df.isnull().sum()
```

```python
#total_cases column has some null value so checking their distribution for filling the null value

df['total_cases'].plot(kind = "kde", figsize = (22,6))
```

```python
# Filling null value with median 

df['total_cases'].fillna(df['total_cases'].median()).plot(kind = "kde", figsize = (22,6))
```

```python
df['total_cases'].describe()
```

```python
df['total_cases'].fillna(df['total_cases'].median(), inplace = True)
```

```python
df.isna().sum()
```

```python
#total_deaths columns filling null value with median

df['total_deaths'].fillna(df['total_deaths'].median()).plot(kind = "kde", figsize = (22,6))
```

```python
#total_deaths columns has some null values so try to verify their distribution

df['total_deaths'].plot(kind = "kde", figsize = (22,6))
```

```python
df['total_deaths'].describe()
```

```python
df['total_deaths'].fillna(df['total_deaths'].median(),inplace = True)
```

```python
df.isna().sum()
```

Filling Null Value in stringency_index column

```python
df[df['stringency_index'].isnull()]['location'].nunique()
```

```python
df['location'].nunique()
```

```python
df['stringency_index'].plot(kind = "kde", figsize = (22,6))
```

```python
df['stringency_index'].fillna(df['stringency_index'].mean()).plot(kind = "kde", figsize = (22,6))
```

```python
df['stringency_index'].fillna(method ='ffill').plot(kind = "kde", figsize = (22,6))
```

```python
df['stringency_index'].fillna(method ='ffill', inplace = True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df[df['gdp_per_capita'].isnull()]['location'].nunique()
```

```python
df[df['gdp_per_capita'].isnull()]['population'].plot(kind = "kde", figsize = (22,6))
```

```python
sim_pop_gdp = df[df['gdp_per_capita'].isnull()]['population'].mean()
sim_gdp_pc = df[(df['population'] > sim_pop_gdp*0.8) & (df['population'] < sim_pop_gdp*1.2 )]['gdp_per_capita'].mean()
```

```python
df['gdp_per_capita'].fillna(sim_gdp_pc, inplace = True)
```

```python
df.isnull().sum()
```

```python
df[df['human_development_index'].isnull()]['location'].nunique()
```

```python
df[df['human_development_index'].isnull()]['population'].plot(kind = "kde", figsize = (22,6))
```

```python
sim_pop_gdp = df[df['human_development_index'].isnull()]['population'].mean()
sim_pop_hdi = df[(df['population'] > sim_pop_gdp*0.8) & (df['population'] < sim_pop_gdp*1.2 )]['human_development_index'].mean()
```

```python
df['human_development_index'].fillna(sim_pop_hdi, inplace = True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.drop(columns = ['Unnmaed:9', 'Unnmaed:10', 'Unnmaed:11',
       'Unnmaed:12', 'Unnmaed:13'], axis = 1, inplace = True)
```

```python
df.columns = ['Code', 'Country', 'date', 'total_cases', 'total_deaths',
       'stringency_index', 'population', 'gdp_per_capita',
       'human_development_index']
```

```python
df.head()
```

# **Top 10 Countries with Highest Covid Cases**

```python
df.groupby("Country").agg({"total_cases" : sum }).sort_values(by = "total_cases", ascending = False).head(10).plot(kind = "bar", figsize = (22,4), color = "r", xlabel = "Country", ylabel = 'total_cases  10^8')
```

# **Total number of deaths among the countries with the highest number of covid-19 cases**

```python
ctdc = df.groupby("Country", as_index = False).agg({"total_cases" : sum , "total_deaths" : sum }).sort_values(by = "total_cases", ascending = False).head(10)
fig,axs = plt.subplots(figsize = (22,10))

plt.subplot(2,1,1)
plt.ylabel("Total_Case 10^8")
plt.bar(ctdc["Country"],ctdc["total_cases"], color= 'y')
plt.title(" Top 10 Country highest number of cases")
plt.subplot(2,1,2)
plt.ylabel("Total_Death 10^8")
plt.bar(ctdc["Country"],ctdc["total_deaths"], color = 'r')
plt.title("Total number of deaths among the countries with the highest number of covid-19 cases ")

plt.show()
```

```python
ax = df.groupby("Country").agg({"total_cases" : sum , "total_deaths" : sum }).sort_values(by = "total_cases", ascending = False).head(10).plot(kind = "bar", figsize = (22,5) ,stacked = True)
ax.set_xlabel("Country")
ax.set_ylabel("total_cases & total_deaths")
```

## **Percentage of total deaths and total cases among all the countries with the highest number of covid-19 cases**

```python
dfl = df.groupby("Country").agg({"total_cases" : sum , "total_deaths" : sum }).sort_values(by = "total_cases", ascending = False)
dfl["Death_percent"] = (dfl["total_deaths"] / dfl["total_cases"])*100
dfl["total_cases_in_%"] = 100- dfl["Death_percent"]
dfl[['Death_percent',"total_cases_in_%"]].head(10).plot(kind = "bar", stacked = True , figsize = (22,5), color = {"y", "r"}, xlabel = "Country", ylabel = "% death of total_cases")
```

# **calculate the death rate of Covid-19 cases**

```python
dfdr = df.groupby("Country").agg({"total_cases" : sum , "total_deaths" : sum }).sort_values(by = "total_cases", ascending = False)
dfdr["death_rate"] = (dfdr["total_deaths"] / dfdr["total_cases"])*100
dfdr.head(10)
```

```python
dfdr["death_rate"].plot(kind = "kde", figsize = (22,6))
```

```python
dfdr["death_rate"].median()
```

## **Imact of Covid 19 on GDP Per capita**

```python
#Sorce WORLD BANK DATA
df3 = pd.read_csv('GDP_percapita.csv')
```

```python
df3.dropna(inplace = True)
```

```python
df3.columns = ['Country Name', 'Code', '2019', '2020']
```

```python
dfgdpc = df3.merge(df, on ="Code")[['Code','2019', '2020','total_cases', 'total_deaths']].groupby('Code', as_index = False).agg({'2019': np.mean, '2020': np.mean,'total_cases':sum, 'total_deaths':sum})
```

```python
#GDP percapita in 2019 before Covid

dfgdpc[["Code","2019"]].set_index("Code").plot(kind = "bar", stacked = True, figsize = (25,5), color = "g")
```

```python
#GDP percapita in 2020 during covid

dfgdpc[["Code","2020"]].set_index("Code").plot(kind = "bar", stacked = True, figsize = (25,5), color = "r")
```

```python
#Comparsion between 2020 and 2019 GDP per capita 

dfgdpc[["Code","2019",'2020']].set_index("Code").plot(kind = "bar", stacked = True, figsize = (25,5))
```
